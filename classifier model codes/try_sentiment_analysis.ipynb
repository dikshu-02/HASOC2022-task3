```python
#Importing traing data
train = pd.read_csv('/content/drive/MyDrive/ML_text_processing/MOLD_SPLIT/subtask_a.tsv','\t',skiprows=1,names=['id','tweet','subtask_a'])
```

```python
Xfeatures = train['tweet']
ylabels = train[['subtask_a']]
```

```python
# Split Data
from sklearn.model_selection import train_test_split
x_traina, x_testa, y_traina, y_testa = train_test_split(Xfeatures, ylabels, test_size=0.3, random_state=0)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('xlm-roberta-base')
```

```python
x_temp = model.encode(np.array(x_traina), batch_size=20,show_progress_bar=True)
x_temptesta = model.encode(np.array(x_testa), batch_size=20,show_progress_bar=True)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
clf = MLPClassifier(hidden_layer_sizes=(256,128),max_iter=300)
clf.fit(x_temp, y_traina)
pred = clf.predict(x_temptesta)
print("f1 score:",f1_score(y_testa, pred, average='weighted'))
print("acc:",accuracy_score(y_testa, pred))
print(classification_report(y_testa, pred))
```
